```python
import pandas as pd

# 교통 소통정보 CSV 불러오기
city_df = pd.read_csv('월별 소통정보 (자치구별-첨두시별).csv', encoding='utf-8')  # 필요 시 encoding='cp949'

# 구조 확인
print(city_df.info())
print(city_df.head(20))

city_df
```

```python
import pandas as pd

highway_df = pd.read_csv('서울시설공단_서울도시고속도로 노선별 요일별 속도_20250507.csv', encoding='cp949')

# 데이터 확인
print(highway_df.info())
print(highway_df.head(20))

highway_df
```

```python
# 기준월을 문자열로 변환 후 datetime 형식으로 변환
city_df['일자'] = pd.to_datetime(city_df['기준월'].astype(str) + '01', format='%Y%m%d')

# 확인
print(city_df[['기준월', '일자']].head())

city_df

# 2) 고속도로 진출입 관련 '구'
target_districts = ['서초구', '강남구', '금천구', '송파구', '강동구']

# 3) 해당 자치구만 필터링
new_city_df = city_df[city_df['권역코드명'].isin(target_districts)].copy()

# 4) 필요한 컬럼만 추출
new_city_df = new_city_df[['일자', '권역코드명', '평균속도']]
new_city_df = new_city_df.rename(columns={'권역코드명': '자치구', '평균속도': '자치구_평균속도'})

# 확인
print(new_city_df.tail(20))


```

```python
# highway_df 의 연도 와 월을 합쳐서 '일자' 생성
highway_df['일자'] = pd.to_datetime(highway_df['년도'].astype(str) + '-' + highway_df['월'].astype(str).str.zfill(2) + '-01')

# # 확인
print(highway_df[['일자']].tail())

highway_df


# # 2) 고속도로 관련 필요한 컬럼들 추출 

# 모든 요일 컬럼 포함
cols = ['일자', '도로명', '평일 평균 속도', '일요일(휴일포함)', '월요일', '화요일', '수요일', '목요일', '금요일', '토요일']
highway_clean = highway_df[cols].copy()

# 컬럼명 깔끔하게 정리
highway_clean.columns = ['일자', '도로명', '평일', '일요일', '월요일', '화요일', '수요일', '목요일', '금요일', '토요일']

highway_clean

```

```python
# 자치구 데이터 정제 (중복 제거)
new_city_df = new_city_df.groupby(['일자', '자치구'], as_index=False)['자치구_평균속도'].mean()

# 병합
merged_df = pd.merge(new_city_df, highway_clean, on='일자', how='inner')

# 병합된 데이터 저장
merged_df.to_csv('병합된_교통속도_데이터.csv', index=False, encoding='utf-8-sig')

# 확인
print(f"총 데이터 수: {len(merged_df)}")

merged_df
```
