```python
   
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """

x = np.linspace(0,10,100)
x
plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """

plt.plot(np.sin(x-1), color='blue')
plt.plot(np.sin(x-2), color='g' )
plt.plot(np.sin(x-3), color='#FF9933')
plt.plot(np.sin(x-4), color=(0.5, 0.2, 1.0))
```

```python
""" 2. 선스타일 """

```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """
plt.plot(np.sin(x-1),'r-')    #solid + red
plt.plot(np.sin(x-2),'c--')   #solid + dashed
plt.plot(np.sin(x-3),'k:')    #black + dotted
plt.plot(np.sin(x-4),'b-.')   #blue + dashdot
```

```python
''' 4. 범례 '''
plt.plot(np.sin(x-1),'r-' , label='First')    #solid + red
plt.plot(np.sin(x-2),'c--', label='Second')   #solid + dashed
plt.plot(np.sin(x-3),'k:', label='Third')    #black + dotted
plt.plot(np.sin(x-4),'b-.', label='Fourth')   #blue + dashdot 

plt.legend(loc=1) #1~4
```
