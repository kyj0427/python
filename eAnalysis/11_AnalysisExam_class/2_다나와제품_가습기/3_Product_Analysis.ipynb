```python
# 예제 6-41 결측값 없애기
chart_data = danawa_data_no_duplicated.dropna(axis = 0)
len(chart_data)
```

```python
# 예제 6-42  각 조건의 최댓값과 평균값 정리

use_time_max_value = int(chart_data['연속시간'].max())
use_time_mean_value = int(chart_data['연속시간'].mean())

use_spraying_max_value = int(chart_data['최대분무량'].max())
use_spraying_mean_value = int(chart_data['최대분무량'].mean())

use_power_max_value = int(chart_data['소비전력'].max())
use_power_mean_value = int(chart_data['소비전력'].mean())
```

```python
print(use_time_max_value)
print(use_time_mean_value)

print(use_spraying_max_value)
print(use_spraying_mean_value)

print(use_power_max_value)
print(use_power_mean_value)
```

### 6.4.4 데이터 시각화

```python
# 예제 6-44 저렴한 제품 기준
chart_data = chart_data.sort_values('가격', ascending=False)
chart_data_selected = chart_data[:20]
chart_data_selected[:20]
```

```python

use_time_max_value = chart_data_selected['연속시간'].max()
use_time_mean_value = chart_data_selected['연속시간'].mean()

spraying_max_value = chart_data_selected['최대분무량'].max()
spraying_mean_value = chart_data_selected['최대분무량'].mean()

plt.figure(figsize=(20, 10))
plt.title("가습기 최대분무량 연속시간 TOP 20")
sns.scatterplot(x = '최대분무량', 
                  y = '연속시간', 
                  size = '가격', 
                  hue = chart_data_selected['회사명'], 
                  data = chart_data_selected, sizes = (100, 2000),
                  legend = False)
plt.plot([60, use_power_max_value], 
          [use_time_mean_value, use_time_mean_value], 
          'r--', 
          lw = 1 )
plt.plot([use_power_mean_value, use_power_mean_value], 
          [20, use_time_max_value], 
          'r--', 
          lw = 1 )
for index, row in chart_data_selected.iterrows():
    x = row['최대분무량']
    y = row['연속시간']
    s = row['제품'].split(' ')[0]
    plt.text(x, y, s, size=20)
plt.show()
```
